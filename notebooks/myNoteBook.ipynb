#### Crear Colecciónes

```python
import subprocess

# races_db
subprocess.run([
    "docker", "exec", "f1_races_db",
    "mongosh", "races_db", "--eval",
    "db.createCollection('races')"
])

# drivers_db
subprocess.run([
    "docker", "exec", "f1_drivers_db",
    "mongosh", "drivers_db", "--eval",
    "db.createCollection('drivers')"
])
subprocess.run([
    "docker", "exec", "f1_drivers_db",
    "mongosh", "drivers_db", "--eval",
    "db.createCollection('driverStandings')"
])

# constructors_db
subprocess.run([
    "docker", "exec", "f1_constructors_db",
    "mongosh", "constructors_db", "--eval",
    "db.createCollection('constructors')"
])
subprocess.run([
    "docker", "exec", "f1_constructors_db",
    "mongosh", "constructors_db", "--eval",
    "db.createCollection('constructorStandings')"
])
subprocess.run([
    "docker", "exec", "f1_constructors_db",
    "mongosh", "constructors_db", "--eval",
    "db.createCollection('constructorResults')"
])

# results_db
subprocess.run([
    "docker", "exec", "f1_results_db",
    "mongosh", "results_db", "--eval",
    "db.createCollection('results')"
])
subprocess.run([
    "docker", "exec", "f1_results_db",
    "mongosh", "results_db", "--eval",
    "db.createCollection('qualifying')"
])
subprocess.run([
    "docker", "exec", "f1_results_db",
    "mongosh", "results_db", "--eval",
    "db.createCollection('lapTimes')"
])
subprocess.run([
    "docker", "exec", "f1_results_db",
    "mongosh", "results_db", "--eval",
    "db.createCollection('pitStops')"
])
```
